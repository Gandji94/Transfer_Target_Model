# Updating Existing TransferMarket Webscrapper

In a previous study, I have created a webscrapper to get the transfer fees from the players over a timespan from 2000-2023. Now, we want to add the 2024 time period as well.

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('..\..\src')
from PY_Class_Def import all_trans_window,trans_value
```

```python
all_trans_window(2024)
```

```python
all_trans_df=all_trans_window(2000)
year_run=[]
for x in range(2001,2025):
    loop_df=all_trans_window(x)
    all_trans_df=pd.concat([all_trans_df,loop_df],ignore_index=True)
    year_run.append(x)
```

```python
all_trans_df
```

```python
def trans_value(x):
    x=str(x)
    if 'Mio' in x and not 'Leihgebühr:' in x and 'Leihe' not in x:
        return int(float(x.split(' ')[0].replace(',','.'))*1000000)
        #return int(float(phase1_mio.split(':')[1].replace(',','.'))*1000000)
    elif 'Tsd' in x and not 'Leihgebühr:' in x and 'Leihe' not in x:
        return int(float(x.split(' ')[0].replace(',','.'))*1000)
    else:
        return 0
```

```python
all_trans_df['Transfer_Fee']=all_trans_df['Transfer_Fee'].apply(trans_value)
```
